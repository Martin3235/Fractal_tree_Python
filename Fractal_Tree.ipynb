```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

```

```python
x = np.array([1,2])
y = np.array([3,4])
al=np.pi/4




def line(axis,prestart,start, iterations): 

    z10 = prestart[0]
    z20 = prestart[1]
    z1 = start[0]
    z2 = start[1]
    length = np.sqrt((z1-z10)**2+(z2-z20)**2)

    xla = np.array([z1,z1+(np.cos(al)*(z1-z10)-np.sin(al)*(z2-z20))*0.8])
    xlb = np.array([z2,z2+(np.sin(al)*(z1-z10)+np.cos(al)*(z2-z20))*0.8])
    
    xra = np.array([z1,z1+(np.cos(-al)*(z1-z10)-np.sin(-al)*(z2-z20))*0.8])
    xrb = np.array([z2,z2+(np.sin(-al)*(z1-z10)+np.cos(-al)*(z2-z20))*0.8])


    ax[axis[0],axis[1]].plot(xla, xlb, color='black', linewidth=0.5) 
    ax[axis[0],axis[1]].plot(xra,xrb, color='black', linewidth=0.5)
    #print(length)
    if length>3*0.8**(iterations):
        line(axis,prestart=start, start = np.array([xla[1],xlb[1]]), iterations=iterations)
        line(axis,prestart=start, start = np.array([xra[1],xrb[1]]), iterations=iterations)
#plt.plot([0,0],[0,1], marker='.',color='black')


fig, ax = plt.subplots(6,2)
fig.set_size_inches(10,20)

line(axis=[0,0],prestart=np.array([0,0]),start=np.array([0,3]), iterations=1)
ax[0,0].set_title('N=1')
ax[0,0].set_ylim(0,15)
ax[0,0].set_xlim(-10,10)
ax[0,0].axis('off')

line(axis=[0,1],prestart=np.array([0,0]),start=np.array([0,3]), iterations=2)
ax[0,1].set_title('N=2')
ax[0,1].set_ylim(0,15)
ax[0,1].set_xlim(-10,10)
ax[0,1].axis('off')


line(axis=[1,0],prestart=np.array([0,0]),start=np.array([0,3]), iterations=3)
ax[1,0].set_title('N=3')
ax[1,0].set_ylim(0,15)
ax[1,0].set_xlim(-10,10)
ax[1,0].axis('off')

line(axis=[1,1],prestart=np.array([0,0]),start=np.array([0,3]), iterations=4)
ax[1,1].set_title('N=4')
ax[1,1].set_ylim(0,15)
ax[1,1].set_xlim(-10,10)
ax[1,1].axis('off')


line(axis=[2,0],prestart=np.array([0,0]),start=np.array([0,3]), iterations=5)
ax[2,0].set_title('N=5')
ax[2,0].set_ylim(0,15)
ax[2,0].set_xlim(-10,10)
ax[2,0].axis('off')


line(axis=[2,1],prestart=np.array([0,0]),start=np.array([0,3]), iterations=6)
ax[2,1].set_title('N=6')
ax[2,1].set_ylim(0,15)
ax[2,1].set_xlim(-10,10)
ax[2,1].axis('off')


line(axis=[3,0],prestart=np.array([0,0]),start=np.array([0,3]), iterations=7)
ax[3,0].set_title('N=7')
ax[3,0].set_ylim(0,15)
ax[3,0].set_xlim(-10,10)
ax[3,0].axis('off')

line(axis=[3,1],prestart=np.array([0,0]),start=np.array([0,3]), iterations=8)
ax[3,1].set_title('N=8')
ax[3,1].set_ylim(0,15)
ax[3,1].set_xlim(-10,10)
ax[3,1].axis('off')


line(axis=[4,0],prestart=np.array([0,0]),start=np.array([0,3]), iterations=9)
ax[4,0].set_title('N=9')
ax[4,0].set_ylim(0,15)
ax[4,0].set_xlim(-10,10)
ax[4,0].axis('off')


line(axis=[4,1],prestart=np.array([0,0]),start=np.array([0,3]), iterations=10)
ax[4,1].set_title('N=10')
ax[4,1].set_ylim(0,15)
ax[4,1].set_xlim(-10,10)
ax[4,1].axis('off')

line(axis=[5,0],prestart=np.array([0,0]),start=np.array([0,3]), iterations=11)
ax[5,0].set_title('N=11')
ax[5,0].set_ylim(0,15)
ax[5,0].set_xlim(-10,10)
ax[5,0].axis('off')


line(axis=[5,1],prestart=np.array([0,0]),start=np.array([0,3]), iterations=12)
ax[5,1].set_title('N=12')
ax[5,1].set_ylim(0,15)
ax[5,1].set_xlim(-10,10)
ax[5,1].axis('off')
```
